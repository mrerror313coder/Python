## Factorial Using Recursion

#### Solving problem using Recursion * Basic Case   * Recursion Case

```python
n = int(input("Enter Number"))

def fact(n):
    if n==0:    # Basic Case
        return 1
    else:
        return n*fact(n-1)   # Recursion Case

if n<0:
    print("Factorial of Negative Number can't be calculated")
elif n==0:
    print(f"Factorial of {n} is 1")
else:
    result=fact(n)    
    print(f"Factorial of {n} is {result}")
```

### Python Program to find power of a given Number :- Use Recursion.

```python
def power(n,p):
    if p==0:
        return 1
    return n*power(n,p-1)

print(power(3,3))
print()
print(power(2,5))
```

### Write a Python Program number is Prime or not (use recursive way)

```python
# Prime Numbers :- 2, 3, 5, 7, 11, 13, 17, .........
n = int(input("Enter Number : "))
def isPrime(n,i):
    if i<0:
        return -1
    if i==1:
        return 1
    if n%i == 0:
        return 0
    return isPrime(n,i-1)


ind = isPrime(n,n-1)
if ind==1:
    print(f"{n} is a Prime Number")
if ind==0:
    print(f"{n} is Not a Prime Number")
if ind==-1:
    print("Negative Number never Prime Number")
```

### Python Program for counting number of digits in given Number. (Use Recursion)

```python
n = int(input("Enter a Number : "))
def count_of(n):
    if n<10:
        return 1
    return 1+count_of(n//10)

print(f"{n} has contain {count_of(n)} digits")
```

## Python Program for Fibonacci Series Using Recursion.

```python
n = int(input("Enter Number of Terms : "))

def fibo(n):
    if n==1:
        return 0
    if n==2:
        return 1
    return (fibo(n-2)+fibo(n-1))
for i in range(1,n+1):
    print(fibo(i),end=" ")
```

## Python Program for SUM of First N-Numbers :- Use Recursion.

```python
def sum1(n):
    if n==1:
        return 1
    return n+sum1(n-1)

n = int(input("Enter Number : "))
print(sum1(n))
```

```python
def sum2(n,result):
    if n==0:
        return result
    result = result+n
    return sum2(n-1,result)

n = int(input("Enter Number : "))
print(sum2(n,result=0))
```
